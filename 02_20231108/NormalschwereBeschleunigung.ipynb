```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import scipy.signal as signal
```

```python
def somigliana(breitengrad):
    gamma_equator = 9.780325
    gamma_pole = 9.832185
    gamma = gamma_equator * (
        1 +
        5.2790414 * 1e-3 * np.sin(breitengrad) ** 2 +
        2.32718 * 1e-5 * np.sin(breitengrad) ** 4 +
        1.262 * 1e-7 * np.sin(breitengrad) ** 6
    )
    return gamma
```

```python
breitengrade = np.arange(-np.pi / 2, np.pi / 2, np.pi / 200)

beschleunigungen = somigliana(breitengrade)
```

```python
fig, ax = plt.subplots(figsize=(30 / 2.54, 15 / 2.54))
ax.plot(
    beschleunigungen,
    breitengrade / np.pi * 180,

    '.-',
)
ax.grid()
ax.set_xlabel(r'Gravitationsbeschleunigung $[m/s^2]$')
ax.set_ylabel('Breitengrad [Grad]')
ax.set_title('Normalschwere nach Somigliana')
fig.savefig('normalschwere_somigliana.jpg', dpi=300)
```
